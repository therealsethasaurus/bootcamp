## Boolean indexing of data frames


```python
df.loc[df['participant number'] == 42, 'percent correct']
```

```python
# can pull whole record for that participant

df.loc[df['participant number'] == 42, :]
```

```python
# pull out all records of females under the age of 21

df.loc[(df['age'] < 21) & (df['gender'] == 'f'), :]
```

```python
# pull out all females under 30 who got ore than 85% of the face matching 
# tests correc

# clearer to se t up Boolean indexing first

inds = (df['age'] < 30) & (df['gender'] == 'f') & (df['percent correct'] > 85)

# take a look
inds
```

```python
df.loc[inds,:]
```

```python
# of interest in Boolean indexing is that we never had to write a looop

# we could have done the following

# Initialize array of Boolean indices
inds = [False] * len(df)

# Iterate over the rows to check if the row should be included
for i, r in df.iterrows():
    if r['age'] < 30 and r['gender'] == 'f' and r['percent correct'] > 85:
        inds[i] = True
        
# make out selection with Boolean indexing
df.loc[inds, :]
```

## Applying functions to Pandas objects


```python
print('Insomniacs:', np.mean(df.loc[df['insomnia'], 'percent correct']))
print('Control:   ', np.mean(df.loc[~df['insomnia'], 'percent correct']))
```

~ operator:
* Bit switcher
* Changes all Trues to Flases and vice versa
* It functions like a logical not

## Styling a data frame


```python
# highlight rows corresponding to women who scored at or above 75%
def highlight_high_scoring_females(s):
    if s['gender'] == 'f' and s['percent correct'] >= 75:
        return ['background-color: #7fc97f'] * len(s)
    else:
        return ['background-color: lightgray'] * len(s)
    
df.head(10).style.apply(highlight_high_scoring_females, axis=1)
```

```python
# Want percent correct column with a bar corresponding to value in the column

# df.syle.bar() 
# subset kwarg specifies which columns are to have bars

df.head(10).style.bar(subset=['percent correct'], vmin=0, vmax=100)
```

```python
df.head(10).style.background_gradient(subset=['percent correct'], cmap='Reds')
```
