# 3.2: Split-Apply-Combine of the frog data set

```python
import numpy as np
import pandas as pd
```

```python
# a
df = pd.read_csv('data/frog_tongue_adhesion.csv', comment='#')
```

```python
df
```

```python
grouped = df.groupby('ID')
```

```python
grouped['impact force (mN)'].std()
```

```python
# b
def coeff_of_var(data):
    """Compute coefficient of variation from an array of data."""
    return np.std(data) / np.mean(data)
```

```python
grouped = df.groupby('ID')
```

```python
df_cv = grouped[['impact force (mN)', 'adhesive force (mN)']].agg(coeff_of_var)
```

```python
df_cv.reset_index()
```
